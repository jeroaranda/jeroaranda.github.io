Este modelo esta basado en un crecimiento exponencial con las tasas de crecimiento presentadas por china como se pueden consultar en este paper 
(https://www.medrxiv.org/content/10.1101/2020.02.07.20021154v1.full.pdf)

es decir si el n√∫mero de infectados que depende del tiempo es ùêº(ùë°) = ùëí^(ùëü(ùë°‚àíùë°0)), y consideramos ùëü =.29, como fue experimentalmente calculada, y alineamos los datos de mexico del primero contagio con el ùë°0 del crecimiento exponencial de China.

Despu√©s los modelos de decrecimiento que comienzan en la supuesta fecha en la que se toman medidas, est√°n dados por un modelo de bounded growth en donde la tasa de decrecimiento es igual que la ùëü anterior es decir .29

Es importante notar que aunque la implementacion de las medidas puede cambiar drasticamente su eficiencia, medidas igual de eficientes generan cambios completamente distintos dada la fecha de su implementaci√≥n y estas crecen a medida que son mas tardadas.



(we assumed exponential growth for the infected population I* in Wuhan, ùêº‚àó(ùë°) = ùëí^ùëü(ùë°‚àíùë°0) , where ùëü is the exponential growth rate and ùë°0 is the time of the exponential growth initiation, i.e. ùêº‚àó(ùë°0) = 1. Note that ùë°0 is likely to be later than the date of the first infection event, because multiple infections may be needed before the onset of exponential growth[...] , the growth rate of the outbreak is estimated to be 0.29 per day (CI: 0.21 to 0.37 per day), a much higher rate than two recent estimates (1, 5). This growth rate corresponds to a doubling time of 2.4 days)

```python
color = plt.cm.magma_r(np.linspace(0.1,0.9,3))
fig,ax=plt.subplots()

"""ax.plot(x,y)
ax.plot(x,y1)
ax.set_ylim((0,30000))
"""
a=.29
x=np.linspace(0,58,200)
y=np.exp(a*x)
ax.plot(x,y,c='#81DCFF',linewidth=3,zorder=1,scaley=True,label='Escenario 1')
esc=['Escenario 2','Escenario 3','Escenario 4']
color=['#50709E','#06316A','#2DAADA']
for j,i in enumerate([20,34,38]):
    tMedida=i
    l1=2*np.exp(a*tMedida)
    k1=(l1-np.exp(a*tMedida))/(l1*np.exp(-a*tMedida))



    
    x1=np.linspace(tMedida,58,200)

    y1=l1*(1-k1*np.exp(-a*x1))


    ax.plot(x1,y1,c=color[j],linewidth=3,label=esc[j])
    




ax.scatter(range(14),C.loc[ C['Country/Region'].isin(['Mexico'])].iloc[:,4:-2].T.values.squeeze()[-14:],zorder=10,color='#2DAADA',label='M√©xico')
ax.plot((0,60),(130000,130000),c='#DA345A',linewidth=3,label='Camas de hospital')
ax.plot((0,60),(np.exp(a*45)*.138,np.exp(a*45)*.138),c='#DA345A',linewidth=3,label='Severe cases')
ax.set_ylim((-2000,400000))
ax.set_facecolor('white')
ax.legend(loc='center left')
fig.savefig('esc.png')

```

```python
from matplotlib import pyplot, dates
df=pd.concat([R.iloc[:,4:-2].sum(),D.iloc[:,4:-2].sum(),C.iloc[:,4:-2].sum()])
df=pd.DataFrame({'ts':df.index, 'q':df.values})
df['dates']=dates.datestr2num(df.ts)
df.ts=pd.to_datetime(df.ts,)
n=R.iloc[:,4:-2].sum().shape[0]
m=D.iloc[:,4:-2].sum().shape[0]
df['status']='Recovered'
df['status'].iloc[n:n+m]='Death'
df['status'].iloc[n+m:]='Confirmed'


ax=sns.lmplot(x='dates',y='q',hue='status',data=df,aspect=1.5,order=2,palette=['#68BE8E','#DA345A','#F5B43E','#06316A'])
ax.set_xticklabels(df.ts,rotation=25)
import matplotlib.dates as mdates
myFmt = mdates.DateFormatter('%d-%m-%y')
ax2=ax.axes
ax2[0,0].xaxis.set_major_formatter(myFmt)
ax2[0,0].set_facecolor('white')
a=ax2[0,0].set_ylabel('Quantities')
fig = ax.savefig('c.png',transparent=True)
```
